There you will find an "official" tutorial in plotting with matplotlib. My intent with this module is not to replicate this tutorial, nor try to cover everything. For that you have the docs. What I want to show is the method which I think offers the most structure of code and control in how you make figures. I also want to share some experience of using matplotlib in the physical sciences.

In general, matplotlib can be divided into two frameworks: functional and object-oriented. The functional framework is heavily inspired by matlab, which can be nice for convertees, but the object-oriented version is more popular, and with good reason. We will use the object-oriented framework.

To begin lets simply import matplotlib.pyplot. It's customary to name the import plt.

Note: familiarity with numpy (numerical python) is generally required.

```python
import matplotlib.pyplot as plt
```

When using object-oriented matplotlib, we are generally dealing with two objects (classes): Figure and Axes. The most direct way to construct these are as follows.

```python
fig = plt.figure()
ax = plt.axes()
```

Running the above code produces an empty figure with unit axes. Try commenting out the ax declaration and see what happens. Python should show <Figure size 432x288 with 0 Axes>, i.e. no plot. Generally, figures have sizes and axes, and axes are what does the plotting. The most transparent way to construct this plot is then as follows.

```python
fig = plt.figure()
ax = fig.add_subplot()
```

Where .add_subplot() is an internal method of the Figure object which returns an Axes object. Generally, a figure may consist of multiple axes, and so we may do this.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)  # (nrows, ncols, index)
ax2 = fig.add_subplot(2, 1, 2)
```

Where, as noted in the comment, we have specified the number of rows and columns, as well as the index. The index tells us where this subplot should be located and starts in the upper left corner and increases to the right. For demonstration.

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)
```

What we should take from this is that a figure generally consists of multiple axes, not one. It's cumbersome to declare all these axes ax1, ax2, ax3, etc. The most useful way (less transparent) to create figure and axes objects are then as follows.

```python
fig, axes = plt.subplots(1, 3)  # (nrows, ncols)
```

This one-liner produces exactly the same as above. If you only want a single subplot, just do this.

```python
fig, ax = plt.subplots(1, 1)
```

This is the way we will construct Figure and Axes objects from now on. Looking back at our previous plot with three columns, we see that the vertical axis tick labels overlap with the previous subplot. It's also tiny. One way to remedy this is as follows.

```python
fig, ax = plt.subplots(1, 3,  figsize=(10, 8))  #figsize : (width, height) in inches
```

Of course, if the y-axis is the same for all subplots, we can do this.

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 8), sharey=True)
```

These are two keyword arguments. You can find plenty more in the docs: https://matplotlib.org/3.3.0/api/_as_gen/matplotlib.pyplot.subplots.html

For completeness, lets make 4 subplots with shared x- and y-axis.

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8), sharey=True, sharex=True)
```

We now want to start plotting something. For that we will construct some data. To do that we will use numpy.

```python
import numpy as np


N = 50  # number of datapoints
x = np.linspace(0, 10, N)
y = np.cos(x)
```

We can now construct the Figure and Axes object as usual. To make a plot, we call on Axes' internal method .plot().

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.plot(x, y)
```

Generally, we want to label our axes.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.plot(x, y)
ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)
```

Our graph is kinda tiny, and we might prefer a different color. Additionally, we might like to have a grid. We can do this.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.plot(x, y, linewidth=3, color="black")
ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)
ax.grid()
```

Data may not actually be continuous, and as oceanographers we might want to invert the y-axis.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.plot(x, y, linewidth=3, color="black", linestyle=" ", marker="o")
ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)
ax.grid()
ax.invert_yaxis()
```

Try changing linewidth. You'll see it doesn't change anything. That is because we have set the line to be invisible. Markers are invisible by default, but shows up when we specify them. To change their size, you would have to use the "markersize" keyword argument. For more keyword arguments for the .plot() method, please see: https://matplotlib.org/3.3.0/api/_as_gen/matplotlib.axes.Axes.plot.html

This type of graph is nothing but a scatter plot, however, and so we might as well do this.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.scatter(x, y, color="black")
ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)
ax.grid()
ax.invert_yaxis()
```

```python
If you wanted to change their size here you would have to use the "s" keyword argument. It's different because the distinction between marker and line isn't made.

Now, we might be very interested in the second local maxima (remember, we have inverted the y-axis), and so we would like to take a closer look at it. We can do this by setting the axis limits.
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.scatter(x, y, color="black")
ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)
ax.grid()
ax.invert_yaxis()
ax.set_ylim(0.75, 1.25)
ax.set_xlim(5, 7)
```

The y-axis is inverted back when setting limits. It's therefore wise to save inverting for last. Actually, a rule of thumb in my experience is to have all .set_* methods before any method not follow that syntax. Such as .invert_yaxis() and .grid(), although for .grid() it doesn't change anything in this case.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.scatter(x, y, color="black")
ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)
ax.set_ylim(0.75, 1.25)
ax.set_xlim(5, 7)

ax.grid()
ax.invert_yaxis()
```

Lets go back to looking at the entire subplot. We can have multiple graphs in a subplot. For this we will construct some more data and call the .scatter() method again. Here is a page on colours: https://matplotlib.org/3.3.0/gallery/color/named_colors.html

```python
y_exp = np.exp(x/2)

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.scatter(x, y, color="black")
ax.scatter(x, y_exp, color="tab:red")  # Usually the colours are prettier if you put "tab:" in front

ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("y(t) [cm]", fontsize=14)

ax.grid()
ax.invert_yaxis()
```

You might want to have a legend. My preferred way to do this is to use the "label" keyword argument in .plot(), .scatter(), etc., and call on .legend()

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.scatter(x, y, color="black", label="cos(x)")
ax.scatter(x, y_exp, color="tab:red", label="exp(x/2)")

ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("log(y(t)) []", fontsize=14)
ax.set_yscale("log")

ax.grid()
ax.invert_yaxis()
ax.legend(frameon=False, fontsize=24, loc="upper right")  # without frameon=False you will get a box around legend
```

These time series have different orders of magnitude, and so we could try remedying this by setting a log scale. Additionally, we might want to have latex legend labels, and so we can call on latex formatting by using r"".

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.scatter(x, y, color="black", label=r"$\cos\,x$")
ax.scatter(x, y_exp, color="tab:red", label=r"$\exp\,\frac{x}{2}$")

ax.set_xlabel("Time [s]", fontsize=14)
ax.set_ylabel("log(y(t)) []", fontsize=14)
ax.set_yscale("log")

ax.grid()
ax.invert_yaxis()
ax.legend(frameon=False, fontsize=24, loc="upper right")
```

However, our oscillatory data has negative values, which aren't shown. Multiple subplots are more appropriate. Since the data has the same time series, we can construct a figure with two rows and two subplots with shared x-axis.

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)  # returns fig, [ax0, ax1]

axes[0].scatter(x, y, color="black", label=r"$\cos\,x$")
axes[0].set_ylabel("y(t) [cm]", fontsize=14)
axes[0].grid()
axes[0].invert_yaxis()

axes[1].scatter(x, y_exp, color="tab:red", label=r"$\exp\,\frac{x}{2}$")
axes[1].set_xlabel("Time [s]", fontsize=14)
axes[1].set_ylabel("log(y_exp(t)) []", fontsize=14)
axes[1].set_yscale("log")
axes[1].grid()
axes[1].invert_yaxis()

fig.legend(frameon=False, fontsize=24, ncol=2, loc="upper center")
```

Where I also called on the Figure.legend() method, which I have found useful when comparing multiple quantities from different time series. Generally you have to do some fine tuning to get the look just right.
It's becoming quite cumbersome to write all of this again. Usually subplots have similar characteristics, and so here is a suggestion on how you may plot even more subplots. The legend is not necessary in this case, so I neglect to add it.

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)  # returns fig, [ax0, ax1]

colors = ["black", "tab:red"]
ylabels = ["y(t) [cm]", "log(y_exp(t)) []"]
time_series = [y, y_exp]

axes[1].set_xlabel("Time [s]", fontsize=14)
for i in range(len(axes)):
    axes[i].scatter(x, time_series[i], color=colors[i])
    axes[i].set_ylabel(ylabels[i], fontsize=14)
    axes[i].grid()
    axes[i].invert_yaxis()
axes[1].set_yscale("log")  # Here I broke the rule of thumb. If I don't the ylims aren't automatically scaled
fig.tight_layout()
```

You can imagine the space you save if you have more than two similar subplots. I also used fig.tight_layout() which is a very convenient general whitespace remover. Try commenting it out and see how it looks. Note, sometimes it gives you weird results, so be vary. Now I'll make a final figure to demonstrate multiple rows and columns.

```python
y_x2 = x**2
y_sincos = np.sin(x) + np.cos(x)

fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex=True)  # returns fig, [[ax0, ax1], [ax2, ax3]]
axes = axes.flatten()  # for easy iteration

colors = ["black", "tab:red", "tab:green", "tab:blue"]
ylabels = ["y(t) [cm]", "log(y_exp(t)) []", "y_x2(t) [cm]", "y_sincos(t) [cm]"]
time_series = [y, y_exp, y_x2, y_sincos]

[ax.set_xlabel("Time [s]", fontsize=14) for ax in axes[-2:]]  
for i in range(len(axes)):
    axes[i].scatter(x, time_series[i], color=colors[i])
    axes[i].set_ylabel(ylabels[i], fontsize=14)
    axes[i].grid()
    axes[i].invert_yaxis()
axes[1].set_yscale("log")
fig.tight_layout()
```

In this case I actually illustrate a weakness of the plt.subplots() function. It might have been a nicer figure if we had grouped the exp() and x^2 functions together with a shared y-axis, and the trigonometric functions with their own y-axis. plt.subplots() only lets us have all the subplots share these properties. We could then do this.

```python
fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2, sharey=ax1)
ax3 = fig.add_subplot(2, 2, 3, sharex=ax1)
ax4 = fig.add_subplot(2, 2, 4, sharex=ax3, sharey=ax3)

axes = [ax1, ax2, ax3, ax4]

colors = ["black", "tab:blue", "tab:red", "tab:green"]
ylabels = ["y(t) [cm]", "log(y_exp(t)) []", "y_x2(t) [cm]", "y_sincos(t) [cm]"]
time_series = [y, y_sincos, y_exp, y_x2]

[ax.set_xlabel("Time [s]", fontsize=14) for ax in axes[-2:]]  
for i in range(len(axes)):
    axes[i].scatter(x, time_series[i], color=colors[i])
    axes[i].set_ylabel(ylabels[i], fontsize=14)
    axes[i].grid()
    axes[i].invert_yaxis()
axes[2].set_yscale("log")
fig.tight_layout()
```

It's important to have in mind all these methods as once, as they are all useful. Note that the tick labels remain with this method of sharing axis, a small but not subtle difference. They can of course be removed by a method Axes.tick_params(), but I will not do this here. The point is that matplotlib lets you customise to your heart's content, and the object-oriented approach especially gives you a lot of control and lets you keep your code structured. For more information, check out: https://matplotlib.org/3.3.0/api/axes_api.html 

I will now demonstrate how you can make a contourf plot with 3D data.

```python
# constructing data

N = 100

x = np.linspace(0, 10, N)
dx = 10

y = x
dy = dx

X, Y = np.meshgrid(x, y)
z = X**2 + Y**2

u = -2*Y
v = 2*X

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

cmap = plt.get_cmap("magma")
c = ax.contourf(X, Y, z,
                cmap=cmap,
                levels=20
               )

ax.quiver(X[::dx, ::dy], Y[::dx, ::dy], u[::dx, ::dy], v[::dx, ::dy])  # Only taking each dx * dy box
ax.set_xlabel("x [m]", fontsize=14)
ax.set_ylabel("y [m]",fontsize=14)
cbar = fig.colorbar(c, ax=ax)
cbar.set_label(r"z [$m^2$]", fontsize=18)
```
